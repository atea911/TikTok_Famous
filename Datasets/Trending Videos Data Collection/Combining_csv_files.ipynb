# Combining the suggested users video data into one big csv

```python
sug1 = pd.read_csv("sug_users_vids1.csv")
sug2 = pd.read_csv("sug_users_vids2.csv")
sug3 = pd.read_csv("sug_users_vids3.csv")
sug4 = pd.read_csv("sug_users_vids4.csv")
```
